```python
import nltk
import string
import pandas as pd
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
```

```python
messages = pd.read_csv(r"F:/datasets/spam.csv",encoding="cp1252")
```

```python
messages.head()
```

```python
messages.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
```

```python
messages.rename(columns={"v1":'label',"v2":'message'},inplace=True)
```

```python
messages.head()
```

```python
length=messages.message.apply(len)
```

```python
length = pd.DataFrame({"count":length})
```

```python
messages = pd.concat([messages,length],axis=1)
```

```python
messages
```

```python
stopwords.words("english")
```

```python
def text_process(mess):
    """
    1.Remove the Punctuation
    2.Remove the stopword
    3.Return the list of class of clean text words
    """
    nopunc = [char for char in mess if char not in string.punctuation]
    nopunc = "".join(nopunc)
    
    return[word for word in nopunc.split() if word not in stopwords.words("english")]
```

```python
messages['message'].apply(text_process)
```

```python
bow_transformer = CountVectorizer(analyzer=text_process).fit(messages['message'])
```

```python
bow_transformer.vocabulary_
```

```python
messages_bow = bow_transformer.transform(messages.message) #converting in Term Document Matrix
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(messages_bow, messages['label'], test_size = 0.20, random_state = 121)
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(X_train, y_train)
```

```python
#Evaluate the model on the test data set
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
pred = classifier.predict(X_test)
print(classification_report(y_test ,pred ))
print('Confusion Matrix: \n', confusion_matrix(y_test,pred))
print()
print('Accuracy: ', accuracy_score(y_test,pred))
```
